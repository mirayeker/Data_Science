## Sorgu 38:

- select c.customer_id,lower(c.first_name),upper(c.last_name),count(p.rental_id)  from public.payment p 
left join rental r ON r.rental_id =p.rental_id 
left join customer c on p.customer_id=c.customer_id   
where r.return_date<=r.rental_date + 5 * INTERVAL '1 day'
group by c.customer_id,c.first_name,c.last_name
having count(p.rental_id)>10
order by c.customer_id 

- Müşteri ad ları küçük harf, soyadları büyük harf olacak şekilde, Kiralam tarihi ile geri dönüş tarihi arası 5 gün olan müşterilerin 10 dan fazla kiralama yaptıkları toplam kiralama sayısını listele

```python
df_customer = pd.read_sql('select customer_id, first_name, last_name from customer', con=engine)
df_payment = pd.read_sql('select rental_id, customer_id from payment', con=engine)
df_rental = pd.read_sql('select rental_id, return_date, rental_date from rental', con=engine)


df_join1 = pd.merge(df_payment, df_rental, on='rental_id', how='left')
df_join2 = pd.merge(df_join1, df_customer, on='customer_id', how='left')

df_join2['last_name'] = df_join2['last_name'].str.upper()
df_join2['first_name'] = df_join2['first_name'].str.lower()

#Kiralama tarihi ile geri dönüş tarihi arası 5 gün 
df_join2['diff_days'] = ((df_join2['return_date'] - df_join2['rental_date']) / np.timedelta64(1, 'D')).round()

df_join2 =df_join2[(df_join2["diff_days"] == 5 ) & (df_join2["rental_id"].count() > 10 )] 
df_join2 = df_join2.groupby(["customer_id", "first_name","last_name"])["rental_id"].agg("count")
df_join2
```

```python
df_customer = pd.read_sql('select customer_id, first_name, last_name  from customer', con=engine)
df_rental = pd.read_sql('select rental_date, customer_id  from rental', con=engine)


df_join1 = pd.merge(df_customer, df_rental, on='customer_id', how='left')
df_join1["sure"] = df_join1.groupby("customer_id")['rental_date'].shift(1).diff()
df_join1["max"] =df_join1.groupby(["customer_id"]).agg({"sure" : "max"})
df_join1["min"] =df_join1.groupby(["customer_id"]).agg({"sure" : "min"})
df_join1 = df_join1[(df_join1.customer_id == 101)]
df_join1 = df_join1.groupby(["customer_id", "first_name", "last_name"]).agg({"sure" : "max", "sure" : "min" })


df_join1

```
